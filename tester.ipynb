```python
def binary_to_term_no_mask(term):
    term_list = list(term)
    
    
    red_term  = ""
    
    for id,term_type in enumerate(term_list):
            red_term += chr(97 + id)
            red_term += "'" if term_type == '0' else ""
    

    return red_term
```

```python
# print(df)

ft = []
fd = []

for r in df1[1:]:
    # print(r)
    # print(r[1:-1])
    a = ''.join(str(item) for item in r[1:-1]) 
    # print(r[-1])
    if r[-1] == '1':
        ft.append(a)
    elif r[-1] != '0':
        fd.append(a)
        



fts = [binary_to_term_no_mask(b) for b in reversed(ft)]
fds = [binary_to_term_no_mask(b) for b in reversed(fd)]

print(f"func_TRUE = {fts}")
print(f"func_DC = {fds}")
```
